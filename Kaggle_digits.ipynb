```python
import pandas as pd
import numpy as np
```

```python
X_train = pd.read_csv('train_dig.csv')
X_test = pd.read_csv('test_dig.csv')
```

```python
X_train.head(3)
```

```python
y = X_train['label']
```

```python
del X_train['label']
```

```python
X_train.head(3)
```

```python
X_test.head(3)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
parameter_grid = {
            'criterion': ['entropy', 'gini'],
            'max_depth': [10, 15, 20, 50],
            'n_estimators': [10, 20, 50]
        }
clf = RandomForestClassifier()
grid_searcher = GridSearchCV(clf, parameter_grid, verbose=2)
grid_searcher.fit(X_train, y)
clf_best = grid_searcher.best_estimator_
 
print('Best params = ', clf_best.get_params())
```

```python
clf_rf = RandomForestClassifier(n_jobs= 1, n_estimators=50, criterion ='gini', min_impurity_split= None, max_depth=50)
```

```python
clf_rf.fit(X_train, y)
```

```python
predicted_rf = clf_rf.predict(X_test)
```
